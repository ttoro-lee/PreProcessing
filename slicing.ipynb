```python
columns = ['Method', 'Path', 'Protocol']

for i in range(len(df)):
    row = df['test'][i]
    row_list = row.split('\n')
    row_data = row_list[0].split(" ")
    for j in range(1, len(row_list)):
        col, ft = row_list[j].split(": ")
        if col not in columns:
            columns.append(col)
        row_data.append(ft)
    print(len(row_data))
    #new_df.loc[i] = row_data
```

```python
print(columns)
```

```python
len(columns)
```

```python
new_df = pd.DataFrame(np.zeros((481, 28)), columns=columns)
```

```python
new_df.head()
```

```python
for i in range(len(new_df)):
    row = df['test'][i]
    row_list = row.split('\n')
    mtd, path, ptc = row_list[0].split(" ")
    new_df['Method'].loc[i] = mtd
    new_df['Path'].loc[i] = path
    new_df['Protocol'].loc[i] = ptc
    for j in range(1, len(row_list)):
        col, ft = row_list[j].split(": ")
        new_df[col].loc[i] = ft
    #new_df.loc[i] = row_data
```
